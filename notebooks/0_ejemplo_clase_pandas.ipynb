## Aporte de Lizeth León: residuos sólidos y cobertura vegetal

Desde mi rol, Pandas puede apoyar el análisis de variables ambientales relacionadas con residuos sólidos y cobertura vegetal.  
Permite organizar datos como cantidad de residuos recolectados, tipo de material, área con vegetación y número de árboles observados.  
Con estas herramientas se pueden calcular totales, porcentajes y comparar puntos de monitoreo para apoyar diagnósticos ambientales.

```python
import pandas as pd

datos_lizeth = {
    "Variable": ["Residuos orgánicos", "Residuos plásticos", "Papel/cartón", "Área vegetal", "Número de árboles"],
    "Valor": [8, 5, 3, 700000, 5200],
    "Unidad": ["kg", "kg", "kg", "m2", "árboles"]
}

df_lizeth = pd.DataFrame(datos_lizeth)

total_residuos = df_lizeth.loc[0, "Valor"] + df_lizeth.loc[1, "Valor"] + df_lizeth.loc[2, "Valor"]

print("Aporte de Lizeth León - Residuos y vegetación")
print("Total de residuos recolectados:", total_residuos, "kg")

df_lizeth
```
