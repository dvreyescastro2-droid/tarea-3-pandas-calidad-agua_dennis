# Reto 3: Funciones estadísticas y nuevas columnas

En este notebook se calculan estadísticas básicas de variables ambientales relacionadas con calidad del agua. Además, se crean nuevas columnas que permiten interpretar mejor los resultados del monitoreo.

```python
import pandas as pd

df = pd.read_csv("../data/4_resultados_agua.csv")

df.head()
```

```python
df["Temp_corregida"] = df["Temp"] + 0.5
df["OD_indice"] = df["OD_promedio"] * 10

df.head()
```

```python
resumen = df.groupby("Punto")[["Temp", "pH", "OD_promedio"]].mean()
resumen
```

## Conclusión

El uso de funciones estadísticas en Pandas permite analizar de manera eficiente la calidad del agua. 
Desde el enfoque de residuos sólidos, estos resultados pueden ayudar a identificar posibles impactos de la contaminación en variables como pH, temperatura y oxígeno disuelto.
