### 💡 Actividad
- Crea al menos dos condiciones (por pH y temperatura).
- Genera una nueva columna con la categoría resultante.

```python
import pandas as pd

datos = pd.read_csv("../data/4_resultados_agua.csv")

datos.head()
```

```python
ph_alto = datos[datos["pH"] > 7.5]
ph_alto
```

```python
temp_alta = datos[datos["Temp"] > 24]
temp_alta
```

```python
def clasificar_calidad(row):
    if row["pH"] >= 6.5 and row["pH"] <= 8.5 and row["OD_promedio"] >= 5:
        return "Buena"
    else:
        return "Regular o mala"

datos["Calidad_agua"] = datos.apply(clasificar_calidad, axis=1)

datos
```

## Interpretación

Se aplicaron condiciones sobre el pH y la temperatura para identificar cambios en la calidad del agua.

Además, se creó una nueva columna llamada "Calidad_agua", que clasifica los datos en función del pH y el oxígeno disuelto.
