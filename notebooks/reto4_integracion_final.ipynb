# Reto 4: Integración final y diagnóstico ambiental

**Dataset:** data/4_resultados_agua.csv  

Este notebook integra los resultados del análisis de calidad del agua para generar un diagnóstico ambiental general. Se relaciona la interpretación con la importancia de la cobertura vegetal en la protección de fuentes hídricas.

```python
import pandas as pd

df = pd.read_csv("../data/4_resultados_agua.csv")

df.head()
```

```python
resumen_sitio = df.groupby("Punto")[["Temp", "pH", "OD_promedio"]].mean()
resumen_sitio
```

```python
diagnostico = resumen_sitio.copy()
diagnostico["interpretacion"] = "La cobertura vegetal protege la calidad del agua"
diagnostico
```
